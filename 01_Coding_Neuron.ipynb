# 2.0 A Layer of Neurons

```python
inputs = [1, 2, 3, 2.5]
weights1 = [0.2, 0.8, -0.5, 1]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]
bias1 = 2
bias2 = 3
bias3 = 0.5
```

### 2.1 Exercise : Given the 4 inputs, 3 neurons which have 4 weights each and 3 biases, calculate the output of the layer.

```python
outputs = [
    inputs[0] * weights1[0] + inputs[1] * weights1[1] + inputs[2] * weights1[2] + inputs[3] * weights1[3] + bias1,
    inputs[0] * weights2[0] + inputs[1] * weights2[1] + inputs[2] * weights2[2] + inputs[3] * weights2[3] + bias2,
    inputs[0] * weights3[0] + inputs[1] * weights3[1] + inputs[2] * weights3[2] + inputs[3] * weights3[3] + bias3
]

outputs
```

```python
inputs = [1, 2, 3, 2.5]
weights = [[0.2, 0.8, -0.5, 1],
[0.5, -0.91, 0.26, -0.5],
[-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]
```

### 2.2 Exercise : Given the 4 inputs, 3 neurons which have 4 weights each and 3 biases, calculate the output of the layer. Modify it so we can use loop to handle dynamic size of inputs and neurons.

```python
layer_outputs = []

for neuron_weights, neuron_bias in zip(weights, biases):
    neuron_output = 0
    for n_input, weight in zip(inputs, neuron_weights):
        neuron_output += n_input * weight
    neuron_output += neuron_bias
    layer_outputs.append(neuron_output)

print(layer_outputs)
```

```python
layer_outputs = [
    sum(n_input * weight for n_input, weight in zip(inputs, neuron_weights)) + neuron_bias 
    for neuron_weights, neuron_bias in zip(weights,biases)
]

layer_outputs
```

# 4.0 Matrix Product



To perform a matrix product, the size of the second dimension of the left matrix must match the
size of the first dimension of the right matrix. For example, if the left matrix has a shape of `(5, 4)`
then the right matrix must match this 4 within the first shape value `(4, 7)`. The shape of the
resulting array is always the first dimension of the left array and the second dimension of the right
array, `(5, 7)`.

# 5.0 Transposition



A row vector is a matrix whose first dimensions size (the number of rows) equals 1 and the
second dimensions size (the number of columns) equals n  the vector size. In other words, its
a 1n array or array of shape (1, n):




A column vector is a matrix where the second dimensions size equals 1, in other words, its an
array of shape (n, 1):




```python
import numpy as np

a = [1, 2, 3]
b = [2, 3, 4]
c = [3, 4, 5]

a = np.array([a]) # make it become matrix by include one more []
# or 
# a = np.expand_dims(np.array(a),axis=0)
print(a.shape) # (1, 3) # 2 dimentional array (matrix)

b = np.array([b]).T # make it become matrix by include one more []
print(b.shape) # (3, 1) # 2 dimentional array (matrix)

c = np.array(c)
print(c.shape) #  (3,) - 1 dimentional array (vector)

np.dot(a, b) 
# matrix dot product matrix, result is also a matrix
# vector dot matrix, result is a list of vector (list of scalar values)
# vector dot vector, result is a scalar
```

# 6.0 A Layer of Neurons & Batch of Data w/ NumPy



Transpose Second Matrix to peform matrix product







```python
import numpy as np

inputs = [[1.0, 2.0, 3.0, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, -0.8]]

weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2.0, 3.0, 0.5]

layer_outputs = np.dot(inputs, np.array(weights).T) + biases

print(layer_outputs)
```
